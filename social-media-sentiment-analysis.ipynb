```python
import pandas as pd
from textblob import TextBlob
import openpyxl

# Load the dataset from an Excel file
file_path = '/kaggle/input/social-media/Social Media.csv'
data = pd.read_csv(file_path)

# Function to classify sentiment using TextBlob
def classify_sentiment(text):
    analysis = TextBlob(text)
    # Classify sentiment
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply the sentiment classification
data['predicted_sentiment'] = data['text'].apply(classify_sentiment)

# Display the first few rows of the dataframe to verify
print(data.head())

# Save the results to a new Excel file
output_file_path = 'social_media_sentiment_analysis_with_predictions.xlsx'
data.to_excel(output_file_path, index=False)

print(f"Sentiment analysis results saved to {output_file_path}")
```
